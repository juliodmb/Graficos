```python
# metricas do skleaning 
# erro absoluto 
# media do quadro de erro 
# raiz do erro medio quadrado 

# MAE mais facil de entender é o erro medio 
# MSE é mais popular que o MAE porque o MSE punira erros maiores o que tende 
# a ser util no mundo real 
# RMSE é ainda mais popular do que MSE porque o RMSE É interpretado 
# nas unidades y 
```

```python
# como calcular essas tres metricas 
from sklearn import metrics
```

```python
print('MAE', metrics.mean_absolute_error(y_test, predcit))
# media de erro de por ligação é de 0.238 
```

```python
print('MSE', metrics.mean_squared_error(y_test, predcit))
# RAIZ unidade esquadratica 
```

```python
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test, predcit)))
```
